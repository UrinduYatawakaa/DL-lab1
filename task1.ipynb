```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
from scipy.stats import pearsonr, spearmanr

exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 array from exponential distribution:\n", exp_array)

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(loc=0, scale=1, size=100000)

plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=50, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=50, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=50, histtype="step", color="green", label="Normal")
plt.axis([0, 5, 0, 1.2])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Plot of Z = X^2 + Y^2')
plt.show()

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
pokemon_data = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = pokemon_data[features]

pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Pearson Correlation Heatmap')
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
