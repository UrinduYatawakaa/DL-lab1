```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')

new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

print("First 20 samples of the new dataframe:")
print(new_df.head(20))

stats = new_df.describe()
print("\nStatistical Summary:")
print(stats)

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, feature in enumerate(features):
    axes[i].hist(new_df[feature], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

sns.pairplot(new_df)
plt.show()

g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, cmap='Blues', fill=True)
plt.show()
```
